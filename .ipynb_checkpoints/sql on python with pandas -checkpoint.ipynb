## 新增一個欄位並且命名

select * , tip / total_bill as tip_rate from tips 

```python
tips['tip_rate']= tips['tip'] / tips['total_bill']
tips
```

```python
tips = tips.assign(tip_rate = tips['tip'] / tips['total_bill'])
```

## join 

```python
df1 = pd.DataFrame({'key':['A','B','C','D'],
                   'value':np.random.randn(4)})
df2 = pd.DataFrame({'key':['B','C','D','E'],
                   'value':np.random.randn(4)})
```

## inner join 

```python
pd.merge(df1 , df2 , on = 'key')
```

```python
# merge() also offers parameters for cases when you’d like to join one DataFrame’s
# column with another DataFrame’s index.
```

```python
indexed_df2 = df2.set_index('key')
pd.merge(df1 , indexed_df2 , left_on = 'key' , right_index = True)
```

## right 

```python
pd.merge(df1 , df2 , on = 'key' , how = 'left')
```

## left  

```python
pd.merge(df1 , df2 , on = 'key' , how = 'right')
```

## outter join 

```python
pd.merge(df1,df2, on ='key' , how = 'outer')
```

##  UNION ALL AND UNION 

```python
df1 = pd.DataFrame({'city':['Chicago', 'San Francisco', 'New York City'],
                   'rank' :range(1,4)})
df2 = pd.DataFrame({'city':['Chicago', 'Boston', 'Los Angeles'] , 
                   'rank':[1,4,5]})
```

```python
# union 
pd.concat([df1,df2])
```

```python
# union all
pd.concat([df1,df2])
```

```python
# union 
pd.concat([df1, df2]).drop_duplicates()
```
